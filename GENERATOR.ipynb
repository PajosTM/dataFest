```python
#ACADEMIC 
import random
import json
import pandas as pd

# Load the student_data.csv
students_df = pd.read_csv("students_data.csv")

# Define courses for each department
courses = {
    "Science": {
        "compulsory": ["Mathematics", "English", "Physics", "Chemistry"],
        "electives": [
            "Biology", "Agricultural Science", "Computer Science", "Civic Education",
            "Fisheries", "Economics", "Data Processing", "Visual Art",
            "Further Mathematics", "Yoruba"
        ]
    },
    "Arts": {
        "compulsory": ["English", "Literature in English", "History", "Government", "Fine Arts"],
        "electives": [
            "Mathematics", "Economics", "Commerce", "Christian Religious Studies",
            "Marketing", "Biology", "Agricultural Science"
        ]
    },
    "Commercial": {
        "compulsory": ["Financial Accounting", "Commerce", "Economics", "English Language", "Mathematics"],
        "electives": [
            "Geography", "Government", "Agricultural Science", "Biology",
            "Yoruba Language", "Further Mathematics", "Religious and Moral Instructions",
            "Civic Education"
        ]
    }
}

# Function to generate subject scores for all terms
def generate_subject_scores_by_term(department):
    # Get the compulsory and elective subjects for the department
    compulsory_subjects = courses[department]["compulsory"]
    electives = courses[department]["electives"]
    
    # Select additional elective subjects to make a total of 7-9 subjects
    chosen_subjects = compulsory_subjects + random.sample(electives, random.randint(3, 5))  # 3-5 electives

    # Create a dictionary to hold term-wise subject scores
    scores_per_term = {}

    # Generate scores for 8 terms (SS1 First Term to SS3 Second Term)
    for term in range(1, 9):
        term_scores = {}
        for subject in chosen_subjects:
            # Performance distribution: 70% average (50-70), 15% exceptional (80-100), 15% below par (30-49)
            performance_type = random.choices(['average', 'exceptional', 'below_par'], [0.5, 0.15, 0.35])[0]
            if performance_type == 'average':
                score = random.randint(50, 70)
            elif performance_type == 'exceptional':
                score = random.randint(80, 100)
            else:
                score = random.randint(30, 49)
            term_scores[subject] = score
        scores_per_term[f"Term_{term}"] = term_scores

    return scores_per_term

# Generate academic record data
academic_records_data = []
for student in students_df.itertuples():
    # Generate the subject scores for each term based on department
    subject_scores_by_term = generate_subject_scores_by_term(student.department)
    
    # Attendance rate can vary across terms, but for simplicity, we'll use a single average attendance rate
    attendance_rate = round(random.uniform(60, 100), 2)
    
    # Append the record: [student_id (foreign key), department (foreign key), subject_scores (JSON), attendance_rate]
    academic_records_data.append([student.student_id, student.department, json.dumps(subject_scores_by_term), attendance_rate])

# Create the Academic Records DataFrame
academic_records_df = pd.DataFrame(academic_records_data, columns=["student_id", "department", "subject_scores", "attendance_rate"])

# Save the dataset to a CSV file
academic_records_df.to_csv("academic_records.csv", index=False)
print('done')
```

```python
import pandas as pd
import random

# Load the exam results and students data
exam_results_df = pd.read_csv("exam_results.csv")
students_df = pd.read_csv("students_data.csv")

# Function to determine marital status based on pass or fail
def assign_marital_status(pass_or_fail):
    if pass_or_fail == 'Pass':
        return random.choices(['Married', 'Divorced'], [0.8, 0.2])[0]
    else:  # For students who failed
        return random.choices(['Divorced', 'Married'], [0.7, 0.3])[0]

# Function to generate parent involvement data
def generate_parent_involvement(student_id, pass_or_fail, involvement_id):
    education_level = random.choices(['None', 'Primary', 'Secondary', 'Tertiary'], [0.1, 0.1, 0.5, 0.3])[0]  # Example distribution
    attendance_at_meetings = pass_or_fail == 'Pass'  # Pass students' parents more likely to attend meetings
    academic_activity_involvement = pass_or_fail == 'Pass'  # Pass students' parents more likely to be involved
    marital_status = assign_marital_status(pass_or_fail)
    return [involvement_id, student_id, education_level, attendance_at_meetings, academic_activity_involvement, marital_status]

# Generate parent involvement data with an auto-incrementing involvement_id
parent_involvement_data = []
for i, exam_result in enumerate(exam_results_df.itertuples(), start=1):
    parent_record = generate_parent_involvement(exam_result.student_id, exam_result.pass_or_fail, i)
    parent_involvement_data.append(parent_record)

# Create the Parent Involvement DataFrame
parent_involvement_df = pd.DataFrame(parent_involvement_data, columns=["involvement_id", "student_id", "education_level", "attendance_at_meetings", "academic_activity_involvement", "marital_status"])

# Display the first few rows of the dataset
print(parent_involvement_df.head())

# Save the dataset to a CSV file
parent_involvement_df.to_csv("parent_involvement.csv", index=False)
print('done')
```

```python
import pandas as pd
import random
import json

# Load the academic records
academic_records_df = pd.read_csv("academic_records.csv")

# Define the courses for each department
courses = {
    "Science": {
        "compulsory": ["Mathematics", "English", "Physics", "Chemistry"],
        "electives": [
            "Biology", "Agricultural Science", "Computer Science", "Civic Education",
            "Fisheries", "Economics", "Data Processing", "Visual Art",
            "Further Mathematics", "Yoruba"
        ]
    },
    "Arts": {
        "compulsory": ["English", "Literature in English", "History", "Government", "Fine Arts"],
        "electives": [
            "Mathematics", "Economics", "Commerce", "Christian Religious Studies",
            "Marketing", "Biology", "Agricultural Science"
        ]
    },
    "Commercial": {
        "compulsory": ["Financial Accounting", "Commerce", "Economics", "English Language", "Mathematics"],
        "electives": [
            "Geography", "Government", "Agricultural Science", "Biology",
            "Yoruba Language", "Further Mathematics", "Religious and Moral Instructions",
            "Civic Education"
        ]
    }
}

# Define pass threshold
PASS_THRESHOLD = 50

# Function to calculate mean scores across all terms
def calculate_mean_scores(subject_scores):
    scores = json.loads(subject_scores)
    subject_totals = {}
    term_count = len(scores)
    
    # Sum scores across all terms for each subject
    for term_scores in scores.values():
        for subject, score in term_scores.items():
            if subject not in subject_totals:
                subject_totals[subject] = 0
            subject_totals[subject] += score
    
    # Calculate the mean for each subject
    mean_scores = {subject: total_score // term_count for subject, total_score in subject_totals.items()}
    return mean_scores

# Function to generate new WASSCE results based on mean scores
def generate_wassce_results(student_row):
    department = student_row['department']
    subject_scores = student_row['subject_scores']
    mean_scores = calculate_mean_scores(subject_scores)
    
    # Store new scores and pass/fail status
    new_scores = {}
    pass_count = 0
    
    # Evaluate compulsory subjects first
    for subject in courses[department]["compulsory"]:
        if subject in mean_scores:
            # Generate a new score based on the mean
            new_score = random.randint(max(0, mean_scores[subject] - 10), min(100, mean_scores[subject] + 10))
            new_scores[subject] = new_score
            if new_score >= PASS_THRESHOLD:
                pass_count += 1
    
    # Evaluate elective subjects next
    for subject in courses[department]["electives"]:
        if subject in mean_scores:
            new_score = random.randint(max(0, mean_scores[subject] - 10), min(100, mean_scores[subject] + 10))
            new_scores[subject] = new_score
            if new_score >= PASS_THRESHOLD:
                pass_count += 1
    
    # Determine final pass or fail based on compulsory subjects first
    final_result = "Pass" if pass_count >= 5 else "Fail"
    
    return new_scores, final_result

# Generate exam results
exam_results_data = []
for index, student_row in academic_records_df.iterrows():
    subject_scores, result = generate_wassce_results(student_row)
    exam_results_data.append([None, student_row['student_id'], 'WASSCE', student_row['department'], json.dumps(subject_scores), result])

# Create the Exam Results DataFrame
exam_results_df = pd.DataFrame(exam_results_data, columns=["result_id", "student_id", "exam_type", "department", "subject_scores", "pass_or_fail"])

# Save the exam results to a CSV file
exam_results_df.to_csv("exam_results.csv", index=False)
print('Exam results generated successfully.')
```
